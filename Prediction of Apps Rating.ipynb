## Data Imputation / Manipulation
#### Filling NULL values with Mean/Median/Mode
For Numerical Values: Median, For Categorical Values : Mode

```python
# Fill Blank Values with Median
def fill_median(column):
    return column.fillna(column.median())
```

```python
dataSet.Rating = dataSet['Rating'].transform(fill_median)
```

```python
#Check Number of Null values after executing 'fill_median'
dataSet.isnull().sum()
```

Find out Mode of Categorical Columns

```python
dataSet['Type'].mode()
```

```python
dataSet['Current Ver'].mode()
```

```python
dataSet['Android Ver'].mode()
```

```python
dataSet['Type'].fillna(str(dataSet['Type'].mode().values[0]), inplace = True)
dataSet['Current Ver'].fillna(str(dataSet['Current Ver'].mode().values[0]), inplace = True)
dataSet['Android Ver'].fillna(str(dataSet['Android Ver'].mode().values[0]), inplace = True)
```

```python
dataSet.isnull().sum()
```

```python
# Converting Price, Reviews, Installs columns from categorical to numerical
dataSet['Price'] = dataSet['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
dataSet['Price'] = dataSet['Price'].apply(lambda x: float(x))
dataSet['Reviews'] = pd.to_numeric(dataSet['Reviews'], errors = 'coerce') # 'Coerce' means ignore errors, if any, ehile converting into numeric
dataSet['Installs'] = dataSet['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
dataSet['Installs'] = dataSet['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
dataSet['Installs'] = dataSet['Installs'].apply(lambda x: float(x))
```

## Data Visualization

```python
grp = dataSet.groupby('Category') # Grouping of Data on the basis of category
x = grp['Rating'].agg(np.mean) # Calculating Mean 'Categorywise' [Grouping]
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,'ro', color='r')
plt.xticks(rotation=90)
plt.title('Category wise Rating')
plt.xlabel('Categories')
plt.ylabel('Rating')
plt.show()
```
